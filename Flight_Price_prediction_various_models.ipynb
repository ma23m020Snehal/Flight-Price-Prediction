```python
def convert_duration_to_hours(duration_str):
    try:
        # Split the duration string into hours and minutes
        parts = duration_str.split()
        hours, minutes = 0, 0

        for part in parts:
            if part.endswith('h'):
                hours = int(part.rstrip('h'))
            elif part.endswith('m'):
                minutes = int(part.rstrip('m'))

        # Convert minutes to decimal hours
        total_hours = hours + minutes / 60
        return total_hours
    except ValueError:
        print(f"Invalid duration format: {duration_str}")
        return None

# Apply the conversion to the entire DataFrame
final_df['duration_hour'] = final_df['Duration'].apply(convert_duration_to_hours)


```

# Seprate the dataset in X and Y columns

```python
X=final_df.drop(columns=['Price'],axis=1)
y=final_df['Price']
```

# Models

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=4)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
```

```python
def predict(ml_model):
    print('Model is: {}'.format(ml_model))
    model= ml_model.fit(X_train,y_train)
    print("Training score: {}".format(model.score(X_train,y_train)))
    predictions = model.predict(X_test)
    print("Predictions are: {}".format(predictions))
    print('\n')
    r2score=r2_score(y_test,predictions)
    print("r2 score is: {}".format(r2score))

    print('MAE:{}'.format(mean_absolute_error(y_test,predictions)))
    print('MSE:{}'.format(mean_squared_error(y_test,predictions)))
    print('RMSE:{}'.format(np.sqrt(mean_squared_error(y_test,predictions))))

    sns.distplot(y_test-predictions)
```

```python
predict(LinearRegression())
```

```python
predict(RandomForestRegressor())
```

```python
predict(KNeighborsRegressor())
```

```python
predict(DecisionTreeRegressor())
```

```python
from sklearn.svm import SVR
predict(SVR())
```

# Hypertunning the model

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {
    'n_estimators' : [100, 120, 150, 180, 200,220,280],
    'max_features':['auto','sqrt'],
    'max_depth':[5,10,15,20,25],
    }
```

```python
rf=RandomForestRegressor()
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,cv=3,verbose=2,n_jobs=-1,)

rf_random.fit(X_train,y_train)

# best parameter
rf_random.best_params_
```
